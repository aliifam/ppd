```python
import pandas as pd  

df_churning = pd.read_csv('BankChurners.csv')

df_churning.head()
```

```python
df_churning.tail()
```

```python
existing_data = df_churning[df_churning['Attrition_Flag'] == 'Existing Customer']
attriting_data = df_churning[df_churning['Attrition_Flag'] == 'Attrited Customer']
```

```python
import matplotlib.pyplot as plt
# melihat jumlah data pada kolom 'Marital_Status' dan menghitung jumlah data yang berbeda
df = df_churning['Marital_Status'].value_counts()
# index semua label data yang berbeda
label = df.index
# membuat pie chart
plt.pie(df, labels = label, autopct='%.2f%%')
plt.title('Marital Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

y = np.array([35, 25, 25, 15])
x = ["Web Programmer", "Data Scientist", "DB Admin", "Manager"]

plt.bar(x, y)
# rotasi label
plt.xticks(rotation=45)
plt.ylabel('Jumlah')
plt.show()
```

```python
y1 = existing_data['Marital_Status'].value_counts()
y2 = attriting_data['Marital_Status'].value_counts()

x1 = np.arange(len(y1.index))
x2 = np.arange(len(y2.index))

bar_width = 0.4

plt.bar(x1, y1, bar_width, label='Existing Customer')
plt.bar(x2+bar_width, y2, bar_width, label='Attrited Customer')

plt.title('Marital Status')
plt.ylabel('Jumlah')
plt.xticks(x2 + bar_width/2, labels=df_churning.Marital_Status.unique())

plt.show()
```

```python
# versi 2 lebih mudah pakai fungsi plot.bar
data1 = existing_data['Marital_Status'].value_counts()
data2 = attriting_data['Marital_Status'].value_counts()

df_new = pd.concat([data1, data2], keys=['existing', 'attrited'], axis=1)
df_new.plot(kind='bar')

plt.title('Marital Status')
plt.ylabel('Jumlah')
plt.show()
```

```python
df_new
```

```python
# stacked bar plot
data1 = existing_data['Marital_Status'].value_counts()
data2 = attriting_data['Marital_Status'].value_counts()
df_new = pd.concat([data1, data2], keys=['existing', 'attrited'], axis=1)

df_new.plot(kind='bar', stacked=True)
plt.title('Marital Status')
plt.ylabel('Jumlah')
plt.show()
```

```python
# membuat dataframe baru, yang menjelaskan presentase dari setiap status pernikahan
df_new['existing_percentage'] = df_new['existing'] / (df_new['existing'] + df_new['attrited']) * 100
df_new['attrited_percentage'] = df_new['attrited'] / (df_new['existing'] + df_new['attrited']) * 100
df_new
```
