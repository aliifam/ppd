```python
#check categorical columns
df_X = df.drop(['car_ID', 'price'], axis=1)
df_y = df['price']
cats = df_X.select_dtypes(include=['object', 'bool']).columns
print(cats)
```

```python
# karena data categorical, maka kita perlu melakukan encoding
# kita akan menggunakan label encoding untuk data categorical
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for cat in cats:
    df_X[cat] = le.fit_transform(df_X[cat])

df_X.head()
```

```python
# cek correlation coefficient
corr = df_X.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# top 3 features yang memiliki korelasi tertinggi dengan price
df_X.corrwith(df_y).sort_values(ascending=False).head(3)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
sns.heatmap(df_X.corr(),annot=True,cmap='coolwarm',mask=np.triu(df_X.corr(),k=1))
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df_X.astype(float).values
y = df_y.astype(float).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
reg = LinearRegression()
reg.fit(X_train, y_train)

print('Coef of determination train:', reg.score(X_train, y_train))
print('Coef of determination test:', reg.score(X_test, y_test))

print('coefficients:', reg.coef_)
print('intercept:', reg.intercept_)
y_pred = reg.predict(X_test)
print('Prediction:', y_pred[:5])
print('Real Value:', y_test[:5])
```

```python
# linear regression model evaluation
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('RMSE:', rmse)
print('R2:', r2)
```

```python
# linear regression model visualization
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.Series(y_test[:50].ravel())
data2 = pd.Series(y_pred[:50].ravel())

df_new = pd.DataFrame({'Real Values': data1, 'Predicted values': data2})
df_new.plot(kind='bar', figsize=(15,3))

plt.title('Linear Regression')
plt.xlabel('Sample i')
plt.ylabel('Car Price')
plt.show()
```

```python
# decision tree regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

X = df_X.astype(float).values
y = df_y.astype(float).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
dt = DecisionTreeRegressor(max_depth=10)
dt.fit(X_train, y_train)

print('Coef of determination train:', dt.score(X_train, y_train))
print('Coef of determination test:', dt.score(X_test, y_test))

y_pred = dt.predict(X_test)
print('prediction:', y_pred[:10])
print('real value:', y_test[:10])
```

```python
# benchmarking decision tree regression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('RMSE:', rmse)
print('R2:', r2)
```

```python
# decision tree regression visualization
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.Series(y_test[:50].ravel())
data2 = pd.Series(y_pred[:50].ravel())

df_new = pd.DataFrame({'Real Values': data1, 'Predicted values': data2})
df_new.plot(kind='bar', figsize=(15,3))

plt.title('Decision Tree Regression')
plt.xlabel('Sample i')
plt.ylabel('Car Price')
plt.show()
```

```python
# random forest regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X = df_X.astype(float).values
y = df_y.astype(float).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

print('Coef of determination train:', rf.score(X_train, y_train))
print('Coef of determination test:', rf.score(X_test, y_test))

y_pred = rf.predict(X_test)
print('prediction:', y_pred[:10])
print('real value:', y_test[:10])
```

```python
# random forest regression benchmarking
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('RMSE:', rmse)
print('R2:', r2)
```

```python
# random forest regression visualization
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.Series(y_test[:50].ravel())
data2 = pd.Series(y_pred[:50].ravel())

df_new = pd.DataFrame({'Real Values': data1, 'Predicted values': data2})
df_new.plot(kind='bar', figsize=(15,3))

plt.title('Random Forest Regression')
plt.xlabel('Sample i')
plt.ylabel('Car Price')
plt.show()
```
